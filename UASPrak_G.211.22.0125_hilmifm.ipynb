B. Atribut Stock Name dan Symbol yang memiliki Dividen/Share 5 tertinggi

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data into a pandas DataFrame
df = pd.read_csv('uas_komgraf.csv',sep=';')

# Sort the DataFrame by Dividend/Share in descending order
df_sorted = df.sort_values(by='Dividend/Share', ascending=False)

# Get the top 5 rows
df_top5 = df_sorted.head(5)

# Create a bar plot using Matplotlib
plt.bar(df_top5['Stock Name'], df_top5['Dividend/Share'])

# Set the plot title and axis labels
plt.title('Top 5 Dividend/Share by Stock Name')
plt.xlabel('Stock Name')
plt.ylabel('Dividend/Share')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

C. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi

```python
# Read data from a CSV file
data = pd.read_csv('stock.csv')

# Convert 'Market Value' column to numeric type
data['Market Value'] = pd.to_numeric(data['Market Value'], errors='coerce')

# Sort the data by Market Value in descending order
sorted_data = data.sort_values('Market Value', ascending=False)

# Select the top 5 stocks with the highest Market Value
top_market_values = sorted_data.head(5)[['Stock Name', 'Symbol']]

# Display the top 5 stocks with their Stock Name and Symbol
for index, row in top_market_values.iterrows():
    print(f"Stock Name: {row['Stock Name']}, Symbol: {row['Symbol']}")
```
